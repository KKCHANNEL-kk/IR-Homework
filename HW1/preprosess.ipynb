```python
with open(r'text\HW1.txt', encoding='utf-8-sig') as f:
    text = f.readlines()
```

```python
term_table = []
for line in text:
    line = line.replace(':', ' ')
    for i in line.lower().split():
        term_table.append(i)
```

```python
term_table = list(set(term_table))
```

```python
term_table.sort()
term_table
```
